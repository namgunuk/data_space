# 11. 원 그래프 (기본) circle graph (basic)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 한글 설정
matplotlib.rcParams['axes.unicode_minus'] = False  # - 부호 표기

import pandas as pd
```

```python
values = [30, 25, 20, 13, 10, 2]

labels = ['python', 'java', 'javascript', 'c#', 'c/c++', 'etc']
```

```python
plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]

labels = ['python', 'java', 'javascript', 'c#', 'c/c++', 'etc']

explode = [0.05] * 6

plt.pie(values, labels = labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]

labels = ['python', 'java', 'javascript', 'c#', 'c/c++', 'etc']

explode = [0.05] * 6
plt.title('freq lang')
plt.pie(values, labels = labels, explode=explode)
plt.legend(loc=(1.2, 0.3), title='lang')
plt.show()
```

# 12. 원 그래프 (심화) circle graph (hard)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 한글 설정
matplotlib.rcParams['axes.unicode_minus'] = False  # - 부호 표기

import pandas as pd
```

```python
values = [30, 25, 20, 13, 10, 2]

labels = ['python', 'java', 'javascript', 'c#', 'c/c++', 'etc']
```

```python
colors = ['b', 'g' , 'r', 'c', 'm', 'y']

explode = [0.05] * 6

plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, 
        explode = explode)
plt.show()
```

```python
wedgeprops={'width':0.8}
plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, 
        explode = explode, wedgeprops=wedgeprops)
plt.show()
```

```python
wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':2}
plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, 
        wedgeprops=wedgeprops)
plt.show()
```

```python
def custom_autopct(pct):
    return ('%.0f%%' % pct) if pct >= 10 else ''

plt.pie(values, labels = labels, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors, 
        wedgeprops=wedgeprops)
plt.show()
```

```python
def custom_autopct(pct):
    return ('%.0f%%' % pct) if pct >= 10 else ''

plt.pie(values, labels = labels, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors, 
        wedgeprops=wedgeprops, pctdistance=.7)
plt.show()
```

# 13. 산점도 그래프 scatter graph

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 한글 설정
matplotlib.rcParams['axes.unicode_minus'] = False  # - 부호 표기

import pandas as pd
```

```python
import pandas as pd

df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('eng')
plt.ylabel('mat')
plt.show()
```

```python
import numpy as np
```

```python
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('eng')
plt.ylabel('mat')
plt.show()
```

```python
sizes = df['학년'] * 300
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('eng')
plt.ylabel('mat')
plt.show()
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=.3)  #c 는 색깔 나누는 카테고리
plt.xlabel('eng')
plt.ylabel('mat')
plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=.3)  #c 는 색깔 나누는 카테고리
plt.xlabel('eng')
plt.ylabel('mat')
plt.colorbar(ticks=[1,2,3], label='학년', shrink=.7, orientation='horizontal')  
# shrink 컬러바 크기 조정, orientation 컬러바 밑으로 표시
plt.show()
```
