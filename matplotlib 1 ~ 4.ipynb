# 2. 축 Axes

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 한글 설정
matplotlib.rcParams['axes.unicode_minus'] = False  # - 부호 표기
x = [1, 2, 3]
y = [2, 3, 5]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.plot(x, y)
plt.xlabel('x', color='red')
plt.ylabel('y', color='#00aa00')
```

```python
plt.plot(x,y)
plt.xticks([1, 2, 3])
plt.yticks([2, 5])
plt.show()
```

# 3. 범례 Legend

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 한글 설정
matplotlib.rcParams['axes.unicode_minus'] = False  # - 부호 표기
x = [1, 2, 3]
y = [2, 3, 5]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, label='data')
plt.legend()
```

```python
plt.plot(x, y, label='data')
plt.legend(loc='upper right')
```

```python
plt.plot(x, y, label='data')
plt.legend(loc='lower right')
```

```python
plt.plot(x, y, label='data')
plt.legend(loc=(0.5,0.5))
```

# 4. 스타일 Styles

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 한글 설정
matplotlib.rcParams['axes.unicode_minus'] = False  # - 부호 표기
x = [1, 2, 3]
y = [2, 3, 5]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, linewidth=5)
```

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='x', markersize=10)
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='black')
```

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, 'ro--')  # 색, 마커, 라인스타일
```

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls='--')  
```

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls='--', alpha=0.3)  
```

```python
plt.figure(figsize=(2, 2))
plt.plot(x, y)
```

```python
plt.figure(figsize=(2, 2), dpi=200)
plt.plot(x, y)
```

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```
