## COVID - 19 백신 종류별 접종 인구

```python
days = [1,2,3]  # 1일, 2일, 3일
az=[2,4,8]      # 단위 만명
pfizer=[5,1,3]
md = [1,2,5]

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, md)
```

```python
plt.figure(figsize=(5,5), dpi=100)

plt.plot(days, az, label='az', marker='v')
plt.plot(days, pfizer, label='pfizer', marker='o', ls='--')
plt.plot(days, md, label='md', marker='s', ls='-.')
plt.legend(ncol=3)
```

# 6. 막대 그래프 (기본) Bar graph (basic)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 한글 설정
matplotlib.rcParams['axes.unicode_minus'] = False  # - 부호 표기
```

```python
labels = ['kang', 'seo', 'jeong']
values = [190, 187, 184]

plt.bar(labels, values, color='r')
```

```python
labels = ['kang', 'seo', 'jeong']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['kang', 'seo', 'jeong']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```python
labels = ['kang', 'seo', 'jeong']
values = [190, 187, 184]
ticks = ['1', '2', '3']

plt.bar(labels, values)
plt.xticks(labels, ticks)
```

# 7. 막대그래프 (심화) bar graph (hard)

```python
labels = ['kang', 'seo', 'jeong']
values = [190, 187, 184]

plt.barh(labels, values)
plt.xlim(180, 190)
```

```python
bar = plt.bar(labels, values)

bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('.')
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)
for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='b')
```

# 9. 누적 막대 그래프 accumulate bar graph

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # 한글 설정
matplotlib.rcParams['axes.unicode_minus'] = False  # - 부호 표기

import pandas as pd

df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'])
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어'])
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'], bottom=df['국어'])
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'])
```

```python
plt.bar(df['이름'], df['국어'], label='Korean')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='English')
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'], label='Mathmathic')

plt.xticks(rotation=45)
plt.legend()
```
